### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
my_btc = 1.2
my_eth = 5.3
```

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

<br>

__Here we print json.dumps to understand the indentation of the dictionary and lists from the extracted API__
<br>

```python
# Fetch current BTC price
BTC_response = requests.get(btc_url).json()
print(json.dumps(BTC_response, indent=4))
```

```python
# Fetch current ETH price
ETH_response = requests.get(eth_url).json()
print(json.dumps(ETH_response, indent=4))
```

<br>

__We pull the prices of both Bitcoin and Etherium in Canadian Dollars as required by calling in the keys from the dictionary in sequential order.__

___The sequence is as follows: `data` -> `1` -> `quotes` -> `CAD` -> `price`___

```python
# Compute current value of my crpto
my_btc_value = BTC_response["data"]["1"]["quotes"]["CAD"]["price"]
my_eth_value = ETH_response["data"]["1027"]["quotes"]["CAD"]["price"]

# Print current crypto wallet balance
print(f"The current value of your {my_btc} BTC is ${my_btc_value:0.2f}")
print(f"The current value of your {my_eth} ETH is ${my_eth_value:0.2f}")
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200 # Bonds
my_spy = 50 # Stocks
```

```python
 # Setup Alpaca API environment
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2"
)
```

```python
# Format current date as ISO format
# The date of coding is 24th of Sept, 2023 - Two days prior is backdated and taken as current date 
current_date = pd.Timestamp("2023-09-22", tz="America/New_York").isoformat()

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to "1Day" for Alpaca API
timeframe = "1Day"

# Get current closing prices for SPY and AGG
df_ticker = alpaca.get_bars(
    tickers,
    timeframe,
    start = current_date,
    end = current_date
).df

# Reorganize the DataFrame
# Separate ticker data
SPY = df_ticker[df_ticker['symbol']=='SPY'].drop('symbol', axis=1)
AGG = df_ticker[df_ticker['symbol']=='AGG'].drop('symbol', axis=1)

# Concatenate the ticker DataFrames
df_ticker = pd.concat([SPY, AGG],axis=1, keys=["SPY", "AGG"])

# Display sample data
df_ticker.head(10)
```

```python
# Fetch the closing prices of APY and AGG
# Adding a float will ensure only closing values are returned and not the whole dictionary object

spy_close_price = float(df_ticker["SPY"]["close"])
agg_close_price = float(df_ticker["AGG"]["close"])

print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
my_spy_value = my_spy * spy_close_price
my_agg_value = my_agg * agg_close_price

# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
household_income = 12000

# Consolidate financial assets data
crypto_assets = round((my_btc_value + my_eth_value), 2)
stock_bonds = my_spy_value + my_agg_value

# Create savings DataFrame

## Create a dictionary to store the data
data = {'Assets': ['Crypto', 'Shares'],
        'Amount': [crypto_assets, stock_bonds]}

## Create a DataFrame
df_savings = pd.DataFrame(data)

# Set the 'Assets' column as the index
df_savings.set_index('Assets', inplace=True)

# Display the savings DataFrame
display(df_savings)
```

```python
# Plot savings pie chart
plt.figure(figsize=(5, 5))
plt.pie(df_savings["Amount"], labels=df_savings.index, autopct='%1.1f%%', startangle=120)
plt.title('Composition of Personal Savings')
plt.legend(df_savings.index, loc="best")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()
```

```python
# Set ideal emergency fund
# An ideal emergency fund should be equal to three times the monthly income
emergency_fund = household_income * 3

# Calculate total amount of savings
total_savings = crypto_assets + stock_bonds

# Validate saving health - Displaying results in Bold using \033[0m 

## Check if total savings are greater than the emergency fund
if total_savings > emergency_fund:
    print("\033[1mCongratulations! You have enough money in your emergency fund.\033[0m")

## Check if total savings are equal to the emergency fund
elif total_savings == emergency_fund:
    print("\033[1mCongratulations! You have reached your financial goal of having an emergency fund.\033[0m")

## If total savings are less than the emergency fund, calculate the difference
else:
    difference = emergency_fund - total_savings
    print(f"\033[1mYou are ${difference:.2f} away from reaching your emergency fund goal.\033[0m")
```

## Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
start_date = pd.Timestamp('2018-09-22', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2023-09-22', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG

# Get current closing prices for SPY and AGG
df_stock_data = alpaca.get_bars(
    tickers,
    timeframe,
    start = start_date,
    end = end_date
).df

# Reorganize the DataFrame
# Separate ticker data
# Ensure to create new data, do not overwirte previous data
SPY1 = df_stock_data[df_stock_data['symbol']=='SPY'].drop('symbol', axis=1)
AGG1 = df_stock_data[df_stock_data['symbol']=='AGG'].drop('symbol', axis=1)

# Concatenate the ticker DataFrames
df_stock_data = pd.concat([SPY1, AGG1],axis=1, keys=["SPY", "AGG"])

# Display sample data
df_stock_data.tail(10)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
MC_thirty_years = MCSimulation(
    portfolio_data = df_stock_data,
    weights = [.60,.40],
    num_simulation = 500,
    num_trading_days = 252*30
)
```

```python
# Printing the simulation input data
MC_thirty_years.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
MC_thirty_years.calc_cumulative_return()
```

```python
# Plot simulation outcomes
line_plot = MC_thirty_years.plot_simulation()

# Save the plot for future usage
plt.savefig("MC_thirty_years_sim_plot.png", bbox_inches="tight")

# Show the plot
plt.show()
```

```python
# Plot probability distribution and confidence intervals
dist_plot = MC_thirty_years.plot_distribution()

# Save the plot for future usage
plt.savefig("MC_thirty_years_dist_plot.png", bbox_inches="tight")

# Show the plot
plt.show()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
stats = MC_thirty_years.summarize_cumulative_return()

# Print summary statistics
print(stats)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
ci_lower = round(stats[8]*initial_investment,2)
ci_upper = round(stats[9]*initial_investment,2)

# Print results
print((f"There is a 95% chance that an initial investment of \033[1m${initial_investment}\033[0m in the portfolio"
                    f" over the next 30 years will end within in the range of"
                    f" \033[1m${ci_lower}\033[0m and \033[1m${ci_upper}\033[0m"))
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment_new = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
ci_lower_new = round(stats[8]*initial_investment_new,2)
ci_upper_new = round(stats[9]*initial_investment_new,2)

# Print results
print((f"There is a 95% chance that an initial investment of \033[1m${initial_investment_new}\033[0m in the portfolio"
                    f" over the next 30 years will end within in the range of"
                    f" \033[1m${ci_lower_new}\033[0m and \033[1m${ci_upper_new}\033[0m"))
```
